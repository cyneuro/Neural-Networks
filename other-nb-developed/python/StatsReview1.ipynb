#### Random Variables
According to Wikipedia, **random variables** are variables whose value depends on outcomes of a random event. They can also be split into two parts, continuous and discrete. Discrete requires they refer to a variable from a finite set (i.e. integers) where continuous values can take any real number and are not bounded by a set.

```python
import numpy as np
import matplotlib.pyplot as plt

"""
This is a probability density function for a random variable. The larger the value more likely we are to find the random variable.
"""

x = np.arange(-5, 5, 0.01)
p = 0.2*np.exp(-(x - 3)**2 / 2)/np.sqrt(2 * np.pi) + \
    0.8*np.exp(-(x + 1)**2 / 2)/np.sqrt(2 * np.pi)

plt.plot(x, p)
plt.xlabel("x")
plt.ylabel("p(x)")
plt.title("Probability Density Function")
plt.tight_layout()
plt.show()
```
